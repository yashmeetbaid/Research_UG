```python
import os
import glob
import cv2
from sklearn.model_selection import train_test_split

def write_images(image_list, out_dir):
    """
    Write images from image_list to the specified output directory.

    Parameters:
    image_list (list): List of image file paths.
    out_dir (str): Directory to write the images to.
    """
    if not os.path.exists(out_dir):
        os.makedirs(out_dir)

    for filepath_old in image_list:
        try:
            image = cv2.imread(filepath_old)
            if image is None:
                print(f"Error reading image {filepath_old}")
                continue
        
            # Generate new file path
            temp = filepath_old.split(os.sep)
            category_dir = temp[-2]  # Assuming the second last component is the category
            filename = temp[-1]

            # Create category directory if it doesn't exist
            new_dir = os.path.join(out_dir, category_dir)
            if not os.path.exists(new_dir):
                os.makedirs(new_dir)

            filepath_new = os.path.join(new_dir, filename)
            print(f"Writing at: {filepath_new}")
            cv2.imwrite(filepath_new, cv2.cvtColor(image, cv2.COLOR_RGB2BGR))

        except Exception as e:
            print(f"Error: {str(e)}")

# Directories
train_data_dir = "C:/Users/yashm/Downloads/Dataset 2.0/Biological/Biological"  # Main Directory containing all image folders
train_out = "C:/Users/yashm/Downloads/Dataset 2.0/training_2.0"  # Path to create new directory for train
validation_out = "C:/Users/yashm/Downloads/Dataset 2.0/validation_2.0"  # Path to create new directory for validation

# Get list of all jpg images in the dataset
rasterlist = glob.glob(os.path.join(train_data_dir, '', '*.jpg'), recursive=True)
#rasterlist = glob.glob(train_data_dir + '/*/*.jpg')

# Split data into training and validation sets
train_samples, validation_samples = train_test_split(rasterlist, test_size=0.2, random_state=42)

# Write images to respective directories
print("Writing Train Images..")
write_images(train_samples, train_out)
print("Train Images done. Writing Validation Images..")
write_images(validation_samples, validation_out)

print("Process Done.")
```
